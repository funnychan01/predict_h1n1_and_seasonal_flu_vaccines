```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt #For visualization
```

```python
processed_data_path = "/Users/jackie/Documents/data_science/projects/predict_h1n1_and_seasonal_flu_vaccines/data/processed/"
processed_file_name = "processed_trainset.csv"
train_set = pd.read_csv(processed_data_path+processed_file_name)
```

```python
raw_data_path = "/Users/jackie/Documents/data_science/projects/predict_h1n1_and_seasonal_flu_vaccines/data/raw/"
train_label_file_name = "Flu_Shot_Learning_Predict_H1N1_and_Seasonal_Flu_Vaccines_-_Training_Labels.csv"
train_label = pd.read_csv(raw_data_path+train_label_file_name)
```

```python
processed_data_path = "/Users/jackie/Documents/data_science/projects/predict_h1n1_and_seasonal_flu_vaccines/data/processed/"
processed_test_file_name = "processed_testset.csv"
test_set = pd.read_csv(processed_data_path+processed_test_file_name)
```

```python
X_train, X_test, y_train, y_test = train_test_split(train_set, train_label, test_size=0.2, random_state=42)
```

```python
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)
```

```python
print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")
```

```python
coefficients = model.coef_
intercept = model.intercept_
print(f"Coefficients: {coefficients}")
print(f"Intercept: {intercept}")
```
